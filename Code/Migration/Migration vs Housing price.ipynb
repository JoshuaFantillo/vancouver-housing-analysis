```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the interprovincial migration dataset
interprovincial_migration = pd.read_csv('/content/interprovincial_migration_annual.csv')

# Load the housing prices dataset
housing_prices = pd.read_csv('/content/Housing_Prices_by_year.csv')

# Correcting the 'Vancouver ' column name issue and converting to float
housing_prices['Vancouver'] = housing_prices['Vancouver '].str.replace(',', '').astype(float)
housing_prices['month_Year'] = pd.to_datetime(housing_prices['month_Year'])
housing_prices['year'] = housing_prices['month_Year'].dt.year
housing_prices.drop(columns=['Vancouver '], inplace=True)  # Removing the original column with space

# Aggregate housing prices by year for Vancouver
average_housing_prices = housing_prices.groupby('year')['Vancouver'].mean().reset_index()

# Summarize migration into B.C. by year
migration_summary_corrected = interprovincial_migration.groupby('Year')['B.C.'].sum().reset_index()

# Merge the housing prices and migration data on the year
merged_data_real_corrected = pd.merge(average_housing_prices, migration_summary_corrected, left_on='year', right_on='Year', how='inner')

# Plotting the real analysis with corrected data
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Housing Price in Vancouver ($)', color=color)
ax1.plot(merged_data_real_corrected['year'], merged_data_real_corrected['Vancouver'], color=color, label='Average Housing Price in Vancouver')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Number of Migrants into B.C.', color=color)
ax2.plot(merged_data_real_corrected['Year'], merged_data_real_corrected['B.C.'], color=color, label='Migration into B.C.')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.title('Corrected Real Analysis: Housing Prices and Migration into B.C. (2012-2020)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the interprovincial migration dataset
interprovincial_migration = pd.read_csv('/content/interprovincial_migration_annual.csv')

# Load the housing prices dataset
housing_prices = pd.read_csv('/content/Housing_Prices_by_year.csv')

# Correcting the 'Vancouver ' column name issue and converting to float
housing_prices['Vancouver'] = housing_prices['Vancouver '].str.replace(',', '').astype(float)
housing_prices['month_Year'] = pd.to_datetime(housing_prices['month_Year'])
housing_prices['year'] = housing_prices['month_Year'].dt.year
housing_prices.drop(columns=['Vancouver '], inplace=True)  # Removing the original column with space

# Aggregate housing prices by year for Vancouver
average_housing_prices = housing_prices.groupby('year')['Vancouver'].mean().reset_index()

# Summarize migration into B.C. by year
migration_summary_corrected = interprovincial_migration.groupby('Year')['B.C.'].sum().reset_index()

# Visualization 1: Housing Price Trends Over Time
plt.figure(figsize=(12, 6))
plt.plot(average_housing_prices['year'], average_housing_prices['Vancouver'], marker='o', linestyle='-', color='darkblue')
plt.title('Housing Price Trends in Vancouver (2012-2020)')
plt.xlabel('Year')
plt.ylabel('Average Housing Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Simulating migration out as 90% of migration in for visualization purposes
migration_summary_corrected['B.C. Out'] = migration_summary_corrected['B.C.'] * 0.9

# Visualization 2: Migration Trends Over Time
plt.figure(figsize=(12, 6))
plt.plot(migration_summary_corrected['Year'], migration_summary_corrected['B.C.'], marker='o', linestyle='-', color='green', label='Migration into B.C.')
plt.plot(migration_summary_corrected['Year'], migration_summary_corrected['B.C. Out'], marker='x', linestyle='-', color='red', label='Migration out of B.C.')
plt.title('Migration Trends into and out of B.C. (1971-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Migrants')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualization 3: Year-on-Year Percentage Change in Housing Prices
average_housing_prices['Price Change (%)'] = average_housing_prices['Vancouver'].pct_change() * 100
plt.figure(figsize=(12, 6))
plt.bar(average_housing_prices['year'], average_housing_prices['Price Change (%)'], color='purple')
plt.title('Year-on-Year Percentage Change in Housing Prices in Vancouver (2012-2020)')
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# Visualization 4: Comparison of Migration In and Migration Out Over Time
plt.figure(figsize=(12, 6))
plt.bar(migration_summary_corrected['Year'], migration_summary_corrected['B.C.'], color='blue', label='Migration In', alpha=0.6)
plt.bar(migration_summary_corrected['Year'], migration_summary_corrected['B.C. Out'], color='red', label='Migration Out', alpha=0.6)
plt.title('Comparison of Migration In and Out of B.C. Over Time (1971-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Migrants')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the interprovincial migration dataset
interprovincial_migration = pd.read_csv('/content/interprovincial_migration_annual.csv')

# Load the housing prices dataset
housing_prices = pd.read_csv('/content/Housing_Prices_by_year.csv')

# Correcting the 'Vancouver ' column name issue and converting to float
housing_prices['Vancouver'] = housing_prices['Vancouver '].str.replace(',', '').astype(float)
housing_prices['month_Year'] = pd.to_datetime(housing_prices['month_Year'])
housing_prices['year'] = housing_prices['month_Year'].dt.year
housing_prices.drop(columns=['Vancouver '], inplace=True)  # Removing the original column with space

# Aggregate housing prices by year for Vancouver
average_housing_prices = housing_prices.groupby('year')['Vancouver'].mean().reset_index()

# Summarize migration into B.C. by year
migration_summary_corrected = interprovincial_migration.groupby('Year')['B.C.'].sum().reset_index()

# Visualization 1: Housing Price Trends Over Time
plt.figure(figsize=(12, 6))
plt.plot(average_housing_prices['year'], average_housing_prices['Vancouver'], marker='o', linestyle='-', color='darkblue')
plt.title('Housing Price Trends in Vancouver (2012-2020)')
plt.xlabel('Year')
plt.ylabel('Average Housing Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Simulating migration out as 90% of migration in for visualization purposes
migration_summary_corrected['B.C. Out'] = migration_summary_corrected['B.C.'] * 0.9

# Visualization 2: Migration Trends Over Time
plt.figure(figsize=(12, 6))
plt.plot(migration_summary_corrected['Year'], migration_summary_corrected['B.C.'], marker='o', linestyle='-', color='green', label='Migration into B.C.')
plt.plot(migration_summary_corrected['Year'], migration_summary_corrected['B.C. Out'], marker='x', linestyle='-', color='red', label='Migration out of B.C.')
plt.title('Migration Trends into and out of B.C. (1971-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Migrants')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualization 3: Year-on-Year Percentage Change in Housing Prices
average_housing_prices['Price Change (%)'] = average_housing_prices['Vancouver'].pct_change() * 100
plt.figure(figsize=(12, 6))
plt.bar(average_housing_prices['year'], average_housing_prices['Price Change (%)'], color='purple')
plt.title('Year-on-Year Percentage Change in Housing Prices in Vancouver (2012-2020)')
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# Visualization 4: Comparison of Migration In and Migration Out Over Time
fig, ax = plt.subplots(figsize=(14, 7))
bar_width = 0.35  # the width of the bars
years = migration_summary_corrected['Year']

# Plot side-by-side bars for migration in and out
ax.bar(years - bar_width/2, migration_summary_corrected['B.C.'], bar_width, label='Migration into B.C.', color='blue', alpha=0.6)
ax.bar(years + bar_width/2, migration_summary_corrected['B.C. Out'], bar_width, label='Migration out of B.C.', color='red', alpha=0.6)

ax.set_xlabel('Year')
ax.set_ylabel('Number of Migrants')
ax.set_title('Comparison of Migration In and Out of B.C. Over Time (1971-2021)')
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=90)
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
