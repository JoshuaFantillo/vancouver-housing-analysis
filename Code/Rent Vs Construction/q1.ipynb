```python
import pandas as pd
```

```python
newconstruct_df = pd.read_csv("data/Metro_new_con_sales_count_raw_uc_sfrcondo_month.csv")
newconstruct_df.head(100)
```

```python
rental_df = pd.read_csv("data/Metro_zori_uc_sfrcondomfr_sm_month.csv")
rental_df.head(100)
```

```python
from matplotlib import pyplot as plt



years = 3
months_to_plot = ["2021-10-31","2022-10-31","2023-10-31"]
f, axes = plt.subplots(1, 3, figsize=(15, 10))
for i in range(years):
    rental_less = rental_df[["RegionID", months_to_plot[i]]]
    newconstruct_less = newconstruct_df[["RegionID", months_to_plot[i]]]
    df_plot = rental_less.merge(newconstruct_less, left_on="RegionID", right_on="RegionID", suffixes=('_rent', '_const'))
    df_plot = df_plot.drop([0])

    axes[i].scatter(df_plot[months_to_plot[i]+'_const'], df_plot[months_to_plot[i]+'_rent'])
    axes[i].set_xlabel('new construction')
    axes[i].set_ylabel('rent index')
    axes[i].set_title(month_to_plot)
```
